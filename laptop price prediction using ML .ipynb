```python
import pandas as pd
import sklearn
from sklearn import linear_model
path = "C:/Users/spoorthi s jain/Desktop/Internship/machine learning/2022projects/Linear_MultipleRegression/Laptoprice/laptop_pricing.csv"

data = pd.read_csv(path)
print(data)
print(data.info())

from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
data['Manufacturer'] = le.fit_transform(data['Manufacturer'])

output=data.drop(['Manufacturer','IntelCore(i-)','IntelCoreGen','processing speed(GHz)','Ram(gb)','HDD(gb)','SSD(gb)','Graphics(gb)','ScreenSize(inch)'],'columns')
inputs=data.drop(['Price'],'columns')
print(inputs)
print(output)
model=linear_model.LinearRegression()
model.fit(inputs,output)
res=model.predict([[3,5,7,2.3,4,1000,0,0,14]])
print(res)
coef=model.coef_
print(coef)
intercept=model.intercept_
print(intercept)
y=(112.06244194*2600)+(23388.88007794*4)+(-3231.7190863*3)+221323.0018654
print(y)
```
